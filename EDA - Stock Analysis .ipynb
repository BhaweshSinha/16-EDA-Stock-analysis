# Inferential Statistics

## Data Visualization

## Hist plot of all the numerical columns in the Dataset

```python
df.columns
```

```python
for i in ['Open', 'High', 'Low', 'Close', 'Adj Close','Volume']:
    plt.title(i)
    sns.histplot(df[i],kde=True)
    plt.show()
```

## Violin plot of all the numerical columns in the Dataset

```python
for i in ['Open', 'High', 'Low', 'Close', 'Adj Close','Volume']:
    plt.title(i)
    sns.violinplot(df[i])
    plt.show()
```

## Overall Pairplot of the Dataset

```python
sns.pairplot(df)
plt.show()
```

## Showing the correlation and the heatmap between the attributes

```python
df2 = df[['Open', 'High', 'Low', 'Close', 'Adj Close','Volume']]
```

```python
df2.corr()
```

```python
sns.heatmap(df2.corr(),annot=True)
plt.show()
```

## Line plot between Date and other attributes

```python
df.columns
```

```python
for i  in ['Open', 'High', 'Low', 'Close', 'Adj Close','Volume']:
    plt.title(f'Date vs {i}')
    plt.figure(figsize=(20,10))
    plt.xticks(rotation=90)
    sns.lineplot(x='Date',y=i,data=df)
```
